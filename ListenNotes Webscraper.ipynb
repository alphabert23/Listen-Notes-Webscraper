```python
# Import necessary libraries
from bs4 import BeautifulSoup
import csv

#change this variable into the name you want for the excel file
csv_name = "PodcastingSmarter"

# Open the saved HTML file from Listen Notes with the appropriate encoding
#change the html path to the path of your downloaded Listen Notes html.
with open('PodcastingSmarter.html', encoding='utf-8') as file:
    soup = BeautifulSoup(file, 'html.parser')

# Find all the podcast details in the HTML file
episode_names = soup.findAll("h3", class_="text-black text-base font-semibold w-full mb-2")
durations = soup.findAll(lambda tag: tag.has_attr('aria-label') and tag['aria-label'].startswith('Audio length'))
dates=soup.findAll("time")
descriptions = soup.findAll("div", class_="text-sm line-clamp-2 ln-wrap-line leading-normal")
links=[]
for episode in episode_names:
    links.append(episode.a.get("href"))

#This section creates a CSV file where we will write the data
file = open(csv_name+'.csv', mode='w', encoding='utf-8', newline='')
writer = csv.writer(file)
writer.writerow(["EPISODE TITLE", "DURATION", "DATE", "DESCRIPTION", "LINK"])

#This loop writes the podcast data into a CSV file for Excel    
for episode, duration, date, description, link in zip(episode_names,durations,dates,descriptions,links):
    description_final = description.text.strip().replace('\n', '')
    writer.writerow([episode.text.strip(),duration.text.strip(),date.text.strip(),description.text.strip().strip(),link])
file.close()
```
